```python
import cv2;
import numpy as np;
from google.colab.patches import cv2_imshow
img = cv2.imread("./1.jpeg")
img = img[:,:,0]
cv2_imshow(img)
img_sh = img.shape
num_q = 60;
r = np.linspace(0,255,num_q)
img_q = np.zeros(img.shape,np.uint8)
arr = [];
```

```python
arr = [];
for i in range(img_sh[0]):
  for j in range(img_sh[1]):
    for a,(b) in enumerate(r):
      arr.append(b)
      if img[i,j]<=arr[a] and img[i,j]>arr[a-1] :
        img_q[i,j] = a

```
